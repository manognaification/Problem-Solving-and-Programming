## PROBLEM 2
## PROBLEM STATEMENT
PLOTTING TIME VS VOLTAGE GRAPH

## CONSTRAINTS

## TEST CASES


```python
import  numpy as np

t=np.arange(0.0,2.0,0.1)
v=1+np.sin(2 * np.pi * t)

plt.xlabel('time')
plt.ylabel('voltage')
plt.grid(True)
plt.plot(t,v)
plt.savefig('tes/figures/time-voltage.jpg')
plt.show()
```

## PROBLEM 3
## PROBLEM STATEMENT
Plot a line graph on income vs year for the state of CALFORNIA


## CONSTRAINTS

## TEST CASES


```python
import pandas as pd
df= pd.read_csv('tes/DataFiles/income.csv')
y = df.values[4]
x= df.columns
x
y
```

```python
years = []
for i in range(3, len(x)):
    years.append(x[i])
years   
income= []
for i in range(3, len(y)):
    income.append(y[i])
income
plt.xlabel('years')
plt.ylabel('income')
plt.plot(years,income)
plt.grid(True)
plt.savefig('tes/figures/years-income.jpg')
plt.show()
```

## PROBLEM 6
### PROBLEM STATEMENT
Generate PIE CHART

### CONSTRAINTS

### TEST CASES



```python
import matplotlib.pyplot as plt

labels = 'HINDI','ENGLISH','TELUGU','BENGALI'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
